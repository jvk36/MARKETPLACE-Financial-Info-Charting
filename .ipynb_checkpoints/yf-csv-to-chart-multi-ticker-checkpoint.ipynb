```python
import csv
import matplotlib.pyplot as plt

# Program outputs a price chart comparing two stock ticker prices over time.
# Read the CSV file exported from https://finance.yahoo.com/quote/RIO/history?p=<ticker> as input

with open('RIO.csv') as csv_file:
    reader = csv.reader(csv_file)
    next(reader)  # Skip the header row
    
    # Parse the data
    dates1 = []
    prices1 = []
    for row in reader:
        date1 = row[0]
        price1 = float(row[5])
        
        dates1.append(date1)
        prices1.append(price1)

with open('BHP.csv') as csv_file:
    reader = csv.reader(csv_file)
    next(reader)  # Skip the header row
    
    # Parse the data
    dates2 = []
    prices2 = []
    for row in reader:
        date2 = row[0]
        price2 = float(row[5])
        
        dates2.append(date2)
        prices2.append(price2)

# Create the plot
plt.figure(figsize=(10, 6))  # Set the size of the chart
plt.plot(dates1, prices1, color='r', label='RIO')
plt.plot(dates2, prices2, color='b', label='BHP')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price Over Time')
plt.xticks(rotation=90)
ax = plt.gca()
count = 0
for label in ax.get_xaxis().get_ticklabels():
    count += 1
    if count < 20:
        label.set_visible(False)
    else:
        count = 0
        
# show legent
plt.legend()

# Show the plot
plt.show()

```
