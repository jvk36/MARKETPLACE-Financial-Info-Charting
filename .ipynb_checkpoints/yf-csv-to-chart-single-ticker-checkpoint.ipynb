```python
import csv
import matplotlib.pyplot as plt

# Program outputs a price chart of a stock ticker over time.
# Read the CSV file exported from https://finance.yahoo.com/quote/RIO/history?p=<ticker> as input

with open('RIO.csv') as csv_file:
    reader = csv.reader(csv_file)
    next(reader)  # Skip the header row
    
    # Parse the data
    dates = []
    prices = []
    for row in reader:
        date = row[0]
        price = float(row[5])
        
        dates.append(date)
        prices.append(price)

# Create the plot
plt.figure(figsize=(10, 6))  # Set the size of the chart
plt.plot(dates, prices, color='r', label='RIO')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price Over Time')
plt.xticks(rotation=90)
ax = plt.gca()
count = 0
for label in ax.get_xaxis().get_ticklabels():
    count += 1
    if count < 20:
        label.set_visible(False)
    else:
        count = 0
        
# show legent
plt.legend()

# Show the plot
plt.show()

```
