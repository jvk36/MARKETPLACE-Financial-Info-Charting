```python
import csv
import matplotlib.pyplot as plt
import numpy as np

# Program outputs a price chart over time.
# Read the CSV file exported from https://finance.yahoo.com/quote/RIO/history?p=<ticker> as input

with open('RIO.csv') as csv_file:
    reader = csv.reader(csv_file)
    next(reader)  # Skip the header row
    
    # Parse the data
    dates = []
    prices = []
    volumes = []
    for row in reader:
        date = row[0]
        price = float(row[5])
        volume = int(row[6])
        
        dates.append(date)
        prices.append(price)
        volumes.append(volume)


# Create the figure and subplots
fig, ax1 = plt.subplots(figsize=(15, 9)) # specify the size of the chart to be bigger than default in inches
ax2 = ax1.twinx()  # Create a second y-axis on the same plot

# Plot the stock price data on the first y-axis
ax1.plot(dates, prices, color='r', label='PRICE')

# Add the volume data to the second y-axis
ax2.bar(dates, volumes, color='b', label = 'VOLUME', alpha=0.5)  # Set the alpha parameter to make the bars semi-transparent

# Set the y-axis limits to match the data
ax1.set_ylim(min(prices), max(prices))
ax2.set_ylim(min(volumes), max(volumes))

# Set the labels and title
ax1.set_xlabel('Date')
ax1.set_ylabel('Price')
ax2.set_ylabel('Volume')
ax1.set_title('Stock Price and Volume Over Time - RIO')

ax1.legend(loc='upper left')
ax2.legend(loc = 'upper right')

# xticks = np.arange(len(prices))
# plt.xticks(xticks[::30], dates[::30], rotation=90)
pos = np.arange(len(prices))
ax1.set_xticks(pos[::30])
ax1.set_xticklabels(np.array(dates)[::30], rotation=90)

# Show the plot
plt.show()

```
